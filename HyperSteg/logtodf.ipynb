```python
f= open("MTM.log","r")

all_list = f.readlines()
y=[]

for a in all_list:
    if "100," in a:
        y.append(a.split(",")[2])

```

```python
import seaborn as sns
import numpy as np
import pandas as pd

print(y)
df = pd.DataFrame(y,index=np.arange(0,len(y)))
print(df)
df = df.rename(columns={df.columns[0]:'mse'}).reset_index()
df.mse = pd.to_numeric(df.mse)
df["step3"] = df.mse 
df["step4"] = df.mse 
df["Dataset"] = "MTM"
df["Status"] = "Uncorrupted"
df["Method"] = "HyperSteg-ImagNet-finetuned"
print(df)

df.to_csv("MTM_Uncorrupted_HyperSteg_Finetuning.csv",index=False)
sns.lineplot(data=df,x="index",y="mse")
```
